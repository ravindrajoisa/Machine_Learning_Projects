```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = cv2.imread("Ravindra_Surathkal_beach.jpg")
print(img.shape)
```

```python
plt.imshow(img)   #BGR Color
plt.show()
```

```python
# convert the image from BGR colorspace to RGB
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# resize the image to reduce size
# img = cv2.resize(img, (382,306))
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
# each pixel is of one data point.
```

```python
X = img.reshape((-1,3))
```

```python
# I want 4 dominant colors from the above image. So, set K = 4
```

```python
k = 4
model = KMeans(n_clusters=k)
```

```python
#train this model
model.fit(X)
```

```python
centroids = model.cluster_centers_
```

```python
print(centroids)
```

```python
colors = np.array(centroids, dtype='uint8')
```

```python
print(colors)
```

```python
# visulisation of the above colors
```

```python
i = 1
for color in colors:
    plt.subplot(1,k,i)  #1 means one row, k columns, plotting will happen at i+1 th position
    #plt.axis("off")   # to remove all the unnecessary data
    i = i+1
    mat = np.zeros((100, 100, 3), dtype = 'uint8')  # creating an empty matrix
    mat[:,:,:] = color    # filling the matrix with the RGB color
    plt.imshow(mat)

plt.show()
```

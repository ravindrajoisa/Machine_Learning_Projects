```python
def fill_age(row):
    age = row[0]
    pclass = row[1]

    if pd.isnull(age):
        if pclass==1:
            return 38.23
        elif pclass ==2:
            return 29.87
        else:
            return 25.14
    else:
        return age
```

```python
titanic_df['Age'] = titanic_df[['Age', 'Pclass']].apply(fill_age, axis=1)
```

```python
titanic_df.isnull().sum()
```

```python
# decision trees decides at each node, and doesn't compute distance unlike logestic, linear regression, KNN.
# hence standardization of data is not necessary.
```

```python
# model building using decision tree
```

```python
X = titanic_df.drop(columns=['Survived'])
y = titanic_df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
# import the decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)   # adding max depth to avoid over fitting . Also, you can use Gini instead of Entropy
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.score(X_train, y_train)
```

```python
# above value of 0.98, this is overfitting, without the use of max_depth
# 0.83 is not overfitting.
```

```python
model.score(X_test, y_test)
```

```python
# note that test score increases after adding max_depth, which means we are doing better, means generalization
```

```python
from sklearn.metrics import classification_report, f1_score
```

```python
print(classification_report(y_test, y_pred))
```

```python
f1_score(y_test, y_pred)
```

```python
# 0.73 - this score is not impressive, so add max_depth in line 69
# 0.76 - is a better score. 
```
